## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("./Resources/charity_data.csv")
application_df.head(50)
```

```python
# Drop the non-beneficial ID columns
application_df.drop(['EIN', 'AFFILIATION', 'ORGANIZATION', 'SPECIAL_CONSIDERATIONS'],axis=1,inplace=True)
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_types_count = application_df['APPLICATION_TYPE'].value_counts()
application_types_count
```

```python
index_list = application_types_count.index
index_list
```

```python
# Choose a cutoff value and create a list of application types to be replaced
app_index_list = application_types_count.index
application_types_to_replace = []

for index in range(len(app_index_list)):
    if application_types_count[index] <700:
        application_types_to_replace.append(app_index_list[index])

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,'Other')

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_count = application_df['CLASSIFICATION'].value_counts()
classification_count
```

```python
# check CLASSIFICATION value counts >1
classification_count.loc[classification_count >1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
class_index_list = classification_count.index
classifications_to_replace = []

for index in range(len(class_index_list)):
    if classification_count[index] <1500:
        classifications_to_replace.append(class_index_list[index])

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,'Other')
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Look at NAME value counts for binning
name_count = application_df['NAME'].value_counts()
name_count
```

```python
# Choose a cutoff value and create a list of names to be replaced
name_index_list = name_count.index
name_to_replace = []

for index in range(len(name_index_list)):
    if name_count[index] <3:
        name_to_replace.append(name_index_list[index])

# Replace in dataframe
for n in name_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(n,'Other')
    
# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
numerical_df =  pd.get_dummies(application_df)
numerical_df
```

```python
# Split our preprocessed data into our features and target arrays
y=numerical_df["IS_SUCCESSFUL"]
X=numerical_df.drop("IS_SUCCESSFUL", axis=1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=len(X_train_scaled[0])))

# More hidden layers
nn.add(tf.keras.layers.Dense(units=30, activation='relu'))
nn.add(tf.keras.layers.Dropout(0.2)) # drop out layer
nn.add(tf.keras.layers.Dense(units=12, activation='relu'))
nn.add(tf.keras.layers.Dense(units=8, activation='tanh'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# A callback that saves the model's weights every 5 epochs
mc = tf.keras.callbacks.ModelCheckpoint('./callbacks_optimized/weights{epoch:08d}.h5', save_weights_only=True, period=5)

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50, callbacks=[mc])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
